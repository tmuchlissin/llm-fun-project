```python
import wget
from pyprojroot import here
from langchain_community.document_loaders import TextLoader, UnstructuredURLLoader, UnstructuredFileLoader
```

```python
# Get data from URL --> file

filename = str(here('data/txt/companyPolicies.txt'))
url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/6JDbUb_L3egv_eOkouY71A.txt'

wget.download(url, out=filename)
print('file downloaded')
```

### **Loader Types — Use Case — Suitable Case**

| Loader                     | Function                                                   | Use Case                                                     |
| -------------------------- | ------------------------------------------------------------ | ------------------------------------------------------------------- |
| **TextLoader**             | Load text from `.txt` file                                   | You already have a **local plain text file**                        |
| **StringLoader**           | Convert Python string into Document                          | Text comes from **code, API response, or variable** (not a file)    |
| **UnstructuredURLLoader**  | Load content from a **URL**                                  | Source exists **online** and you want to **skip downloading**       |
| **UnstructuredFileLoader** | Load **any document type** (`PDF, DOCX, PPTX, HTML, TXT...`) | You have **local files in mixed formats**                           |
| **DirectoryLoader**        | Load **multiple files from folder**                          | You have **dataset stored in a directory**                          |
| **PyPDFLoader**            | Extract PDF **page by page**                                 | You need **high-quality PDF parsing** with **page number metadata** |


```python
# Data loader from url (UnstructuredURLLoader)

url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/XVnuuEg94sAE4S_xAsGxBA.txt'

loader = UnstructuredURLLoader(urls=[url])
documents = loader.load()

for i, doc in enumerate(documents):
    print(f"--- Document {i} ---")
    print("page_content:", doc.page_content)
    print("metadata:", doc.metadata)
```

```python
# Data loader from file (txt) (UnstructuredFileLoader)

loader = UnstructuredFileLoader(filename)
documents = loader.load()

for i, doc in enumerate(documents):
    print(f"--- Document {i} ---")
    print("page_content:", doc.page_content)
    print("metadata:", doc.metadata)
```

```python
# Data loader from file (txt)

loader = TextLoader(filename)
documents = loader.load()

for i, doc in enumerate(documents):
    print(f"--- Document {i} ---")
    print("page_content:", doc.page_content)
    print("metadata:", doc.metadata)
```
