```python
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.vectorstores.pgvector import PGVector
from langchain_community.storage import LocalFileStore
from langchain_community.retrievers import ParentDocumentRetriever
from langchain.schema import Document


parent_splitter = CharacterTextSplitter(chunk_size=2000, chunk_overlap=50, separator="\n")
child_splitter = CharacterTextSplitter(chunk_size=400, chunk_overlap=20, separator="\n")

# ---------------------------------------------------------------
# 2ï¸âƒ£ Embeddings (local Ollama model)
# ---------------------------------------------------------------
embedding_function = OllamaEmbeddings(model="bge-m3:latest")

# ---------------------------------------------------------------
# 3ï¸âƒ£ PostgreSQL + PGVector setup
# Pastikan PGVector extension sudah aktif
# ---------------------------------------------------------------
connection_string = "postgresql+psycopg://postgres:password@localhost:5432/rag_db"

vectorstore = PGVector(
    connection_string=connection_string,
    collection_name="split_parents_pgvector",
    embedding_function=embedding_function,
)

# ---------------------------------------------------------------
# 4ï¸âƒ£ Parent Store (Local)
# Menyimpan dokumen besar (parent chunks)
# ---------------------------------------------------------------
store = LocalFileStore("./parent_docs")

# ---------------------------------------------------------------
# 5ï¸âƒ£ Buat ParentDocumentRetriever
# ---------------------------------------------------------------
retriever = ParentDocumentRetriever(
    vectorstore=vectorstore,
    docstore=store,
    child_splitter=child_splitter,
    parent_splitter=parent_splitter,
)

# ---------------------------------------------------------------
# 6ï¸âƒ£ Tambahkan dokumen (sekali saja untuk indexing)
# ---------------------------------------------------------------
docs = [
    Document(page_content="LangChain is a framework for building applications with LLMs."),
    Document(page_content="Retrievers help LangChain fetch relevant data for user queries."),
]

retriever.add_documents(docs)

print("âœ… Dokumentasi berhasil di-ingest ke PGVector.")

# ---------------------------------------------------------------
# 7ï¸âƒ£ Query dokumen
# ---------------------------------------------------------------
query = "What is LangChain retriever?"
results = retriever.get_relevant_documents(query)

for i, doc in enumerate(results, 1):
    print(f"\n--- ðŸ§© Document {i} ---\n{doc.page_content[:400]}")
```
