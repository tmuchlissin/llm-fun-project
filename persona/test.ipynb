```python
import os
from dotenv import load_dotenv
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from langchain_core.runnables import RunnableLambda, RunnableSequence
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
load_dotenv()
GEMINI_API_KEY = os.getenv("GEMINI_API_KEY")

if not GEMINI_API_KEY:
    raise ValueError("⚠️ Harap tambahkan GEMINI_API_KEY ke dalam file .env kamu.")
```

```python
llm =  ChatGoogleGenerativeAI(
        model="gemini-2.5-flash",
        google_api_key=os.getenv('GEMINI_API_KEY'),
        temperature=0,
        max_output_tokens=4000,
        )
```

```python
role = """
Kamu adalah AIWA — Asisten AI Warna Warni Media,
seorang sales properti digital yang sangat ramah, komunikatif, dan profesional.
Kamu membantu calon pembeli menemukan properti sesuai kebutuhan:
rumah, ruko, tanah, apartemen, atau gudang.
Selalu gunakan bahasa yang mudah dipahami dan singkat.
"""

tone = "tegas dan penuh amarah"

template = """
{role}

Pertanyaan calon pembeli: {question}

Gaya berbicara: {tone}

Berikan jawaban yang jelas, menarik, dan informatif.
Jika perlu, ajukan pertanyaan lanjutan untuk memperjelas kebutuhan calon pembeli.
"""

prompt = PromptTemplate.from_template(template)
```

```python
def format_prompt(variables):
    return prompt.format(**variables)

bot_chain = RunnableSequence(
    RunnableLambda(format_prompt),
    llm,
    StrOutputParser(),
)
```

```python
while True:
    query = input("Question: ")
    print(f"Question: ", query)
    if query.lower() in ["quit", "exit", "bye"]:
        print("Answer: Goodbye!")
        break
        
    response = bot_chain.invoke({"role": role, "question": query, "tone": tone})
    print(f"Answer: ", response)
```
